```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

x = np.linspace(0.01, 1, 400)
y = 6 / x**2

colors = cm.viridis(np.linspace(0, 1, len(x)))

plt.figure(figsize=(8, 1.5))

plt.scatter(x, y, c=colors, s=1)

highlight_x = [2**-0, 2**-1, 2**-2, 2**-3, 2**-4]
highlight_y = [6 / (val**2) for val in highlight_x]
plt.scatter(highlight_x, highlight_y, color='grey', zorder=2, linewidths=0.1)

for (hx, hy) in zip(highlight_x, highlight_y):
    x_label = f'$2^{{{int(np.log2(hx))}}}$' 
    y_label = f'{int(hy)}'
    plt.text(hx, hy, f'({x_label}, {y_label})', fontsize=12, color='black', ha='left')

plt.xlabel(r'$\rho$', fontsize=14)
plt.ylabel('RE', fontsize=14)
plt.xlim(0, 1.1)

plt.yscale('log')

plt.grid(True)

plt.legend()

plt.show()
```
